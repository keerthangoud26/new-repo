```python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function
def quadratic(x, a=-1, b=0, c=10):
    return a * x**2 + b * x + c

# Define the stages of the Waterfall Model
stages = ["Requirements", "Design", "Implementation", "Verification", "Maintenance"]

# Equally spaced x positions for stages
x_positions = np.linspace(-2, 2, len(stages))
y_positions = quadratic(x_positions)

# Plot the quadratic curve
x_curve = np.linspace(-2.5, 2.5, 500)
y_curve = quadratic(x_curve)

plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve, linestyle="--", label="Quadratic Curve")

# Add markers and connect them with vertical lines to simulate waterfall
for i in range(len(stages)):
    if i > 0:
        # Horizontal segment
        plt.plot(
            [x_positions[i-1], x_positions[i]],
            [y_positions[i-1], y_positions[i-1]],
            linewidth=2
        )
        # Vertical drop
        plt.plot(
            [x_positions[i], x_positions[i]],
            [y_positions[i-1], y_positions[i]],
            linewidth=2
        )

    # Markers and labels
    plt.scatter(x_positions[i], y_positions[i], color="red", zorder=5)
    plt.text(
        x_positions[i],
        y_positions[i] + 0.3,
        stages[i],
        ha="center",
        fontsize=10
    )

# Styling the plot
plt.title("Waterfall Model Representation Using Quadratic Curve", fontsize=14)
plt.xlabel("Stages", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.axhline(0, linewidth=0.5, linestyle="--")
plt.axvline(0, linewidth=0.5, linestyle="--")
plt.grid(linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend()
plt.show()
```
