```python
# @title Version 1: Hardcoding Variables
# Hardcoded variables
a = 10
b = 20
result = a + b

print(f"Hardcoded Addition: {a} + {b} = {result}")
```

```python
# Version 1: Hardcoded Variables
def calculate_sum_hardcoded():
    # Predefined values
    a = 10
    b = 20

    # Calculation
    result = a + b

    print(f"Variable a: {a}")
    print(f"Variable b: {b}")
    print(f"Sum (Hardcoded): {result}")

# Run the function
calculate_sum_hardcoded()
```

```python
# Version 2: Keyboard Input
try:
    x = int(input("Enter the first number: "))
    y = int(input("Enter the second number: "))
    result = x + y
    print(f"User Input Addition: {x} + {y} = {result}")
except ValueError:
    print("Please enter valid integers!")
```

```python
# Version 2: Keyboard Input with Exception Handling
def calculate_sum_input():
    print("--- User Input Calculation ---")
    try:
        # Taking input from user
        val1 = input("Enter first number: ")
        val2 = input("Enter second number: ")

        # Converting to integers
        num1 = int(val1)
        num2 = int(val2)

        result = num1 + num2
        print(f"The sum is: {result}")

    except ValueError:
        # Handling the error if inputs are not numbers
        print("Error: Invalid input. Please enter numeric values only.")

# Run the function
calculate_sum_input()
```

```python
# @title Version 3: Read From File (Single Input)

# Step A: Create the file (for demonstration purposes in Colab)
with open("single_input.txt", "w") as f:
    f.write("42")

# Step B: Read and process the file
def read_single_input():
    try:
        with open("single_input.txt", "r") as f:
            # Read and clean the input
            content = f.read().strip()
            num = int(content)

            result = num * 2
            print(f"Single Input Read from File: {num} * 2 = {result}")

    except FileNotFoundError:
        print("File not found!")
    except ValueError:
        print("Error in reading or converting the input!")

if __name__ == "__main__":
    read_single_input()
```

```python
# Version 3: Single File Input
import os

def process_single_file():
    filename = 'single_input.txt'

    # Step 1: Create the file (for demonstration purposes)
    with open(filename, 'w') as f:
        f.write("69")
    print(f"File '{filename}' created successfully.")

    # Step 2: Read and Process
    try:
        with open(filename, 'r') as f:
            content = f.read().strip()
            number = int(content)

            # Perform operation (e.g., multiply by 2)
            result = number * 2
            print(f"Read value: {number}")
            print(f"Result (Value * 2): {result}")

    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except ValueError:
        print("Error: The file content is not a valid number.")

# Run the function
process_single_file()
```

```python
# Version 4: Read From File (Multiple Inputs)

# 1. Create the file first
with open("multiple_inputs.txt", "w") as f:
    f.write("3\n5\n7")

# 2. Read the file
try:
    with open("multiple_inputs.txt", "r") as f:
        numbers = [int(line.strip()) for line in f.readlines()]
        result = sum(numbers)
        print(f"Multiple Inputs Read from File: {numbers}, Sum = {result}")
except FileNotFoundError:
    print("File not found!")
except ValueError:
    print("Error in reading or converting the inputs!")
```

```python
# Version 4: Multiple File Inputs
def process_multiple_inputs():
    filename = 'multiple_inputs.txt'

    # Step 1: Create the file with multiple lines
    with open(filename, 'w') as f:
        f.write("10\n20\n30")
    print(f"File '{filename}' created with multiple values.")

    # Step 2: Read, Process, and Sum
    try:
        with open(filename, 'r') as f:
            # List comprehension to read lines and convert to int
            numbers = [int(line.strip()) for line in f.readlines()]

            total = sum(numbers)

            print(f"Numbers read: {numbers}")
            print(f"Total Sum: {total}")

    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")
    except ValueError:
        print("Error: One or more lines in the file are not valid integers.")

# Run the function
process_multiple_inputs()
```

```python
def save_calculation(x, y, filename="output.txt"):
    # Perform calculations
    sum_val = x + y
    prod_val = x * y

    # Save to file
    try:
        with open(filename, "w") as file:
            file.write(f"Sum: {sum_val}\n")
            file.write(f"Product: {prod_val}\n")
        print(f"Success: Data written to {filename}")
    except IOError:
        print("Error: Could not write to file.")

# Execute the function
save_calculation(15, 5)
```
